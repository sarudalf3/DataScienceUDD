## <font color='green'>Tarea 1: Challenging</font>  
Crea una función que tome un número del 1 al 10 y que devuelva la tabla de multiplicación de dicho número (hasta x 10).<br>
Verifica que el númeor ingresado sea mayor que 0 y menor que 11, en caso contrario despliega un mensaje.
Nombra la función como **tabla_mult**

El resultado debería verse como
```python
tabla_mult(3)
```
 1 x 3 =   3 <br> 
 2 x 3 =   6 <br>
 3 x 3 =   9 <br>
 4 x 3 =  12 <br>
 5 x 3 =  15 <br>
 6 x 3 =  18 <br>
 7 x 3 =  21 <br>
 8 x 3 =  24 <br>
 9 x 3 =  27 <br>
10 x 3 =  30 <br>

````python
def tabla_mult(num):
    if num > 0 and num < 11: #Condición rango de entrada
        for i in range(11):
            print(f"{i} x {num} = {i*num}")
    else:
        print("Debe ser número entre 1 y 10")        
````

````python
tabla_mult(5)
````

````python
tabla_mult(15)
````

## <font color='green'>Tarea 3:</font> 

Crea una función que tome una cadena y devuelva el primer carácter de esa cadena en mayúsculas y el resto en minúsculas.<br>
Nombra la función como **first_upper**

Tip:

Busca en internet cómo concatenar strings en Python

````python
def first_upper(txt):
    return txt[0].upper()+txt[1:].lower()
````

<font color="darkorange">Otra forma de hacerlo:</font>

````python
def first_upper2(txt):
    return ''.join([letter.upper() if i==0 else letter.lower() for i, letter in enumerate(txt)])
````

## <font color='green'>Tarea 4:</font> 

Crea una función que tome una cadena y devuelva los dos últimos caracteres. Si hay menos de dos caracteres, devuelve la palabra: "Error".<br>

Tip:

````python
def last_two(txt):
    if len(txt) < 2:
        return 'Error'
    else:
        return txt[-2:]    
````

## <font color='green'>Tarea 5: Challenging</font> 
Dada una lista de enteros, devuelve **True** si la secuencia [1, 2, 3] está en algún lugar de la lista.
<br>
Nombra la función como **seq_check**

Tip:

Use *slicing* y un bucle **for**.

````python
def seq_check(lst):
    for i in range(len(lst)-2):
        if lst[i]==1 and lst[i+1]==2 and lst[i+2]==3:
            return True
            break  #solo para asegurarse de salir del for loop
    return False #para chequear que no encuentra la secuencia en la lista
````

````python
seq_check([1, 2, 3])
````

````python
seq_check([7, 7, 7, 1, 2, 3, 7, 7, 7])
````

````python
seq_check([3, 2, 1, 3, 2, 1, 1, 1, 2, 2, 3, 3, 3])
````

<font color="darkorange">Otra forma de hacerlo:</font>

````python
def seq_check2(lista, lista_ref=[1, 2, 3]):
    result = False
    recorrido = len(lista) - len(lista_ref) + 1
    for i in range(recorrido):
        result = result or lista[i:i+len(lista_ref)] == lista_ref
    return result
````

````python
seq_check2([1, 2, 3])
````

````python
seq_check2([7, 7, 7, 1, 2, 3, 7, 7, 7])
````

````python
seq_check2([3, 2, 1, 3, 2, 1, 1, 1, 2, 2, 3, 3, 3])
````

<font color="darkorange">Otra forma de hacerlo:</font>

````python
def seq_check3(lista):
    secuencia = [1, 2, 3]
    for i in range(0, len(lista)-2):
        if secuencia == lista[i:i+3]:
            return True
    return False
````

````python
seq_check3([1, 2, 3])
````

````python
seq_check3([7, 7, 7, 1, 2, 3, 7, 7, 7])
````

````python
seq_check3([3, 2, 1, 3, 2, 1, 1, 1, 2, 2, 3, 3, 3])
````

## <font color='green'>Tarea 7:</font> 

Dada una lista de enteros, si la longitud de la lista es un número par, devuelve la suma de la lista. Si la longitud de la lista es impar, devuelva el valor máximo en esa lista.

Nombra tu función como **sum_or_max**

````python
def sum_or_max(lst):
    if len(lst) % 2 == 0:
        return sum(lst) #suma los elementos de la lista
    else:
        return max(lst) #encuentra el vaor mayor de la lista
````

````python
sum_or_max([1,2,3])
````

````python
sum_or_max([0,1,2,3])
````

## <font color='green'>Tarea 8: Super challenging</font> 

Los agentes en el campo a veces necesitan hablar en código. Cree una función que tome un nombre (un string, por ejemplo, "Juan Antonio", "Alejandra", etc.) y haga lo siguiente:
* Reemplace todas las vocales con la letra x. Para nuestros propósitos, consideremos estas letras como vocales: a, e, i, o, u, y sus respectivas mayúsculas. 
* Cambie la posición de la primera y última letra.
* Entregue el resultado en mayúsculas.

Tip:
1. Esta tarea es desafiante, divídela en varias partes.
2. AL trabajar sobre una lista, necesitarás unirla en un único string para devolver el resultado. Utiliza para ello el método **.join()** de los strings.

Nombre tu función como **replace_and_switch**

````python
def replace_and_switch(name):
    if type(name) is list:
        name = " ".join(name) #en caso de lista, deja la variable name como texto separadas por " "

    name = name.lower() #Deja los caracteres en minúscula
    name = name.replace('a', 'x').replace('e', 'x').replace('i', 'x').replace('o', 'x').replace('u', 'x') #reemplaza las vocales por x
    name = name[-1] + name[1:-1] + name[0]  #se reordena el texto, la última letra al comienza y la primera letra al final
    name = name.upper() #Deja los caracteres en mayúscula
    return name
````

````python
replace_and_switch("an nb")
````

````python
replace_and_switch('Juan Antonio')
````

````python
replace_and_switch("Marcela")
````

````python
replace_and_switch(["javier","gonzalo"])
````

<font color="darkorange">Otra forma de hacerlo:</font>

````python
def replace_and_switch2(nombre, letras=["a", "e", "i", "o", "u"], reemplazo="X"):
    if type(nombre) is list:
        nombre = " ".join(nombre)
    letras_m = [i.upper() for i in letras]
    for letra in (letras + letras_m):
        nombre = nombre.replace(letra, reemplazo)
    nombre = nombre[-1] + nombre[1:-1] + nombre[0]
    return nombre.upper()
````

````python
replace_and_switch2("an nb")
````

````python
replace_and_switch2('Juan Antonio')
````

````python
replace_and_switch2("Marcela")
````

````python
replace_and_switch2(["javier","gonzalo"])
````
