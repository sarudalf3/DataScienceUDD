2. Con ciclo While

````python
ruta = '/content/drive/MyDrive/Colab Python/01. Fundamentos de Python/Exam 01/' #Definir path
with open (ruta + 'exam.txt','r') as f:  #leer archivo example.txt usando with
  while True: #while loop
    line = f.readline() #readline() lee cada linea del archivo en particular, el formato de salida es texto 
    if line:
      print(type(line))
      print(line)
    else:
      break
f.close()
````

3. Con método .read()

````python
ruta = '/content/drive/MyDrive/Colab Python/01. Fundamentos de Python/Exam 01/' #Definir path
with open (ruta + 'exam.txt','r') as f:  #leer archivo example.txt usando with
	lines = f.read() #read() lee todo el archivo y lo almacena como variable de cadena
	print(type(lines))
	print(lines)
f.close()
````
