## <font color='green'>Tarea 1: Challenging</font> 
Cree una clase Esfera que acepte un radio en la instanciación y tenga métodos de volumen y área de superficie.

Tip:

* Utiliza la función **round( )** para ajustar los decimales de salida 

```python
import math #Importar valor pi
```

```python
class Esfera ():
    
    def __init__ (self, radio):
        self.radio = radio
    
    def area_superficie(self):
        return f'La superficie de la esfera de radio {self.radio} es {round(4 * math.pi * self.radio**2, 2)}'
    
    def volumen(self):
        return f'El volumen de la esfera de radio {self.radio} es {round((4/3) * math.pi * self.radio**3, 2)}'
```

```python
s = Esfera(1)
```

```python
s.area_superficie()
```

```python
s.volumen()
```

## <font color='green'>Tarea 2: Super Challenging</font> 

**Este ejercicio lo desarrollaremos en varios pasos.**
1. Crea una clase llamada *GuessingGame*.
2. La clase no necesita ningún argumento proporcionado por el usuario para la instanciación.
3. En el método __init__ (el constructor), haz que la clase establezca un atributo *self.rand_choice* en un número entero aleatorio entre 0-10.
4. Crea un método en la clase llamado *reset_random* que vuelva a generar un número aleatorio (también entre 0-10) para el atributo *self.rand_choice*
5. Crea un método llamado *guess* (adivinar) que use la función *input( )* para aceptar un número del usuario. Este método debe imprimir una declaración que indique si la suposición del usuario fue correcta o no. 

Bonus track: agregua una lógica en el método *guess* que informe al usuario cuando su número está por arriba o por debajo del número a adivinar.

Tip:
* En el punto 3 use *import random*
* Te dejamos el esqueleto de la clase

```python
import random as rd

class GuessingGame():
    
    def __init__(self):
        self.rand_choice = rd.randint(0,10)       
        
    def reset_random(self):
        self.rand_choice = rd.randint(0,10)
   
    def guess(self):
        while True:
            numusuario = input('Ingrese un número entre 0 y 10: ')
            # Confirma si el número ingresado es un entero, sino lo pide nuevamente
            if numusuario.isdigit() == False:
                print(f'Debe ingresar un número entero \n')
                continue
            #Si el número es entero y está entre 0 y 10 compara con el número aleatorio a adivinar    
            elif int(numusuario) in range(11):
                if self.rand_choice < int(numusuario):
                    print(f'Su número {numusuario} es mayor al número a adivinar \n')
                    continue
                elif self.rand_choice > int(numusuario):
                    print(f'Su número {numusuario} es menor al número a adivinar \n')
                    continue
                else:
                    print(f'Ha acertado el número a adivinar era {self.rand_choice} \n')
                    break
            #Si el número es entero, pero está fuera del rango lo vuelve a pedir.
            else:
                print(f'Ingresó un numero fuera de rango \n')
                continue
```

```python
g = GuessingGame() #Declara variable g de la clase GuessingGame
```

```python
g.rand_choice #Nmro a adivinar
```

```python
g.reset_random() #Cambia el nmro a adivinar
```

```python
g.guess() #Método para adivinar el número aleatorio
```
