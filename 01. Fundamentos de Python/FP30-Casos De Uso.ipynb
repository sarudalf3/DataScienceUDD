```python
def numprim(valor):
    """
    Determina si numero es primo

    Param: valor (numero a evaluar)

    Output: mensaje indicando si es o no es primo
    """    
    for i in range(2, valor):
         print(i)
         if valor%i == 0:
             print("No es numero primo")
             return
    print("Es numero primo")
```

## Casos de uso: Validación de emails (usando expresiones regulares)

Hay librerias para validar el email, aunque en esta oportunidad se implementará una función para su validación. 

[Librería Python para validar e-mail](https://pypi.org/project/py3-validate-email/)

[Información útil para mejoras en el código](https://mailtrap.io/blog/python-validate-email/)

```python
import re #Importar expresiones regulares

regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b' #Expresión regular para email

def validar_email(email):
    if(re.fullmatch(regex, email)):
        print("Email válido")
        return True
    else:
        print("Email inválido")
        return False
```

# Juego de dados 🎲🎲

Simulación de juego de dados donde compites con el computador...

Lanzas 2 dados de manera aleatoria y si tus dados suman más que los de la máquina ganas 🎉🎉🎉

```python
import random
import time

def lanza_dados(n):
    """
    Función devolver un n enteros aleatorios de rango 1 a 6
    Inputs: 
        n: largo de la lista de salida
        num1: numero de dado 2
    Salidas: 
        lista con n enteros aleatorios
    """
    return [random.randrange(6) + 1 for i in range(n)]

def posicion_punto(n, sim):
    """
    Función que devuelve lista de entrada para imprimir un dado
    Inputs: 
        n: número entero de 1 a 6, representa el dado a imprimir.
        sim: simbolo que se imprime en los dados 
    Salidas: 
        Lista con los str que se imprime en el dado (9 str)
    """
    if n == 1:
        return [" ", " ", " ", " ", sim, " ", " ", " ", " "]
    if n == 2:
        return [sim, " ", " ", " ", " ", " ", " ", " ", sim]
    if n == 3:
        return [sim, " ", " ", " ", sim, " ", " ", " ", sim]
    if n == 4:
        return [sim, " ", sim, " ", " ", " ", sim, " ", sim]
    if n == 5:
        return [sim, " ", sim, " ", sim, " ", sim, " ", sim]
    if n == 6:
        return [sim, " ", sim, sim, " ", sim, sim, " ", sim]

def mostrar_dados(num1, num2):
    """
    Función para imprimir resultados de lanzamiento de dados
    Inputs: 
        num1: numero de dado 1
        num1: numero de dado 2
    Salidas: 
        Representación de par de dados en pantalla de acuerdo a num1 y num2
    """

    sim = "\033[91m●\033[0m" # agrega color rojo al punto    
    x = posicion_punto(num1, sim) # dado 1
    y = posicion_punto(num2, sim) # dado 2

    fila1 = f"+---+---+---+  +---+---+---+\n"
    fila2 = f"| {x[0]} | {x[1]} | {x[2]} |  | {y[0]} | {y[1]} | {y[2]} |\n"
    fila3 = f"+---+---+---+  +---+---+---+\n"
    fila4 = f"| {x[3]} | {x[4]} | {x[5]} |  | {y[3]} | {y[4]} | {y[5]} |\n"
    fila5 = f"+---+---+---+  +---+---+---+\n"
    fila6 = f"| {x[6]} | {x[7]} | {x[8]} |  | {y[6]} | {y[7]} | {y[8]} |\n"
    fila7 = f"+---+---+---+  +---+---+---+\n"
    return fila1 + fila2 + fila3 + fila4 + fila5 + fila6 + fila7

lanzamiento_jugador = lanza_dados(2)
lanzamiento_pc = lanza_dados(2)

nombre = input("Ingrese su nombre: ")
nombre = f"\033[1;94m{nombre}\033[0m"

print(f"\nLanzamiento {nombre}...")
time.sleep(1)
print(mostrar_dados(lanzamiento_jugador[0], lanzamiento_jugador[1]))

time.sleep(1)
print("Lanzamiento \033[1;92mMáquina\033[0m...")
time.sleep(1)
print(mostrar_dados(lanzamiento_pc[0], lanzamiento_pc[1]))

time.sleep(1)
if sum(lanzamiento_jugador) > sum(lanzamiento_pc):
    print(f"🎊🎉🎊 {nombre} ganaste 💰💰💰")
else:
    print(f"Cueck!!!! perdiste {nombre} 😭😭😭, inténtalo nuevamente...")
```

#Precios Commodities
En la mayoría de las industrias productivas es de interés conocer los precios de los Commodities (metales, combustibles, etc.), tanto los precios actuales u observados en un rango acotado, asi como también su evolución histórica. Desde la página web del Banco Mundial se pueden obtener en formato Excel los precios históricos de los Commodities desde 1960 a la fecha, en resolución mensual/anual.

**Se propone una función que permita acceder al archivo Excel actualizado del Banco Mundial, donde el usuario pueda seleccionar un Commodity y acceder a los precios actualizados del año en curso (formato tabla).**

#Código

```python
import pandas as pd

url="https://thedocs.worldbank.org/en/doc/5d903e848db1d1b83e0ec8f744e55570-0350012021/related/CMO-Historical-Data-Monthly.xlsx"

def commodities (commodity):
    """
    Función para obtener precio de un commodity dentro del año en curso
    Inputs: commodity seleccionado tipo string
    Salidas: Tabla con commodity con precios mensuales de año en curso"
    """

    from datetime import date # Importar librería datetime para acceder a fecha de hoy

    dicc={'Petróleo Crudo':1,'Carbón':5,'Gas Natural':7,'Cobre':64,'Oro':69} #Diccionario para acceder a partir de nombre de commodity a # columna de Excel
    index=dicc[commodity] #Obtener índice columna Excel con nombre Commodity
    
    #Abrir archivo Excel desde página web. Por facilidad se usa Líbrería Pandas (DataFrame)

    xls = pd.ExcelFile(url) #Abrir Excel con Url
    datos=pd.read_excel(xls, sheet_name='Monthly Prices',header=4) #Leer tabla con precios de commodities

    #Dar Formato a Dataframe

    datos=datos.rename(columns={datos.columns[index]:commodity+" "+datos.iloc[0,index]}) #Cambiar nombre Inglés a Español a Commodity seleccionado
    datos=datos.drop(0) #Eliminar primeras dos filas con basura
    datos=datos.drop(1)

    datos['Año']=datos[datos.columns[0]].str.slice(0, 4).astype(int) #Crear y dar formato a columna Año
    datos['Mes']=datos[datos.columns[0]].str.slice(5, 7).astype(int) #Crear y dar formato a columna Mes

    datos[datos.columns[index]]=datos[datos.columns[index]].astype(float) #Convertir datos a float (puede que los lea como string en pd.read_excel())
    datos=datos[['Año','Mes',datos.columns[index]]] #Filtrar selección

    datos=datos.loc[datos['Año']==date.today().year] #Filtrar año en curso
    datos.set_index(['Año', 'Mes'], inplace=True) #setear como index Año,Mes

    return datos
```

```python
#Programa de prueba

while True:
    try:
        comm=input('Escriba Commodity: (Petróleo Crudo, Carbón, Gas Natural, Cobre, Oro) ')
        tabla=commodities(comm)
        break
    except:
        print('Asegurese de escribir correctamente el Commodity!')
print('Precios en doláres (US)')
tabla
```

### Juego del ahorcado

El jugador debe adivinar la palabra secreta.

Se tiene un cierto número de intentos, y en caso de acertar una letra se imprime en el lugar que corresponde.

```python
import random

turnos = 10
print (f'Hola, éste es el juego del ahorcado. Tienes {turnos} turnos')
print ("")

# Vamos a definir las palabras posibles y la palabra a adivinar
Listado = ["hola", "chao", "arco", "area"]
Palabra = random.choice(Listado)

lista_palabra = list (Palabra)
lista_palabra2 =[]

for item in lista_palabra:
    lista_palabra2.append('_')

#Ahora realizamos el loop hasta que no queden turnos

while turnos > 0:
    
    print(' '.join(lista_palabra2))
    letra = input('Dame una letra: ')

    error = False

    if letra not in lista_palabra:
        # Ha fallado
        error = True
        turnos -= 1

        print ("Error") 

        print (f' Tienes {turnos} turnos disponibles')

    else:
        # Acierta, ingresamos la letra a la lista
        for key, value in enumerate(lista_palabra):
            if value == letra:
                lista_palabra2[key] = value

    if (turnos == 0):
        print (f' Perdiste, la palabra era {Palabra}') 
    elif lista_palabra == lista_palabra2:
        print (f' Ganaste, la palabra era {Palabra}')
        break
```
