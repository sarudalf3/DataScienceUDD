```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#mount drive files root
from google.colab import drive
drive.mount('/content/drive')
```

```python
#read database (both years)
pop_2002 = pd.read_excel("/content/drive/MyDrive/Colab_Python/03. Análisis Exploratorio y Estadística/Archivos/tabulados_censo_nacional.xlsx",
              sheet_name="Cuadro 1", skiprows=4, usecols="B:G", nrows=22)
pop_2002.drop(index=0, inplace=True)

pop_2017 = pd.read_excel("/content/drive/MyDrive/Colab_Python/03. Análisis Exploratorio y Estadística/Archivos/tabulados_censo_nacional.xlsx",
              sheet_name="Cuadro 1", skiprows=4, usecols="I:N", nrows=22)
pop_2017.drop(index=0, inplace=True)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
age = ['0-4','5-9','10-14','15-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-64','65-69',
       '70-74','75-79','80-84','85-89','90-94','95-99','+100']

pop_2002['Hombres'] = pop_2002['Hombres']  
pop_2002['Mujeres'] = - pop_2002['Mujeres']
pop_2002['Edad'] = age

pop_2017['Hombres'] = pop_2017['Hombres.2']  
pop_2017['Mujeres'] = - pop_2017['Mujeres.2']
pop_2017['Edad'] = age

pop_2002.head()
```

```python
pop_2017.head()
```

```python
AgeClass = ['100+','95-99','90-94','85-89','80-84','75-79','70-74','65-69','60-64','55-59','50-54','45-49','40-44',
            '35-39','30-34','25-29','20-24','15-19','10-14','5-9','0-4']

labels = ['80K', '60K', '40K', '20K', '0', '20K', '40K', '60K', '80K']

fig, axes = plt.subplots(1, 2, figsize=(20,8))
#Gráfico año 2002
bar_plot = sns.barplot(x='Hombres', y='Edad', data=pop_2002, order=AgeClass, orient='h', ax=axes[0], color='royalblue', label="Hombres", lw=0)
bar_plot = sns.barplot(x='Mujeres', y='Edad', data=pop_2002, order=AgeClass, orient='h', ax=axes[0], color='tomato', label="Mujeres", lw=0)
axes[0].set_xlim(-800000,800000)
bar_plot.set_title("Pirámide problacional año 2002")
bar_plot.set_xticklabels(labels)
bar_plot.set_xlabel('')
bar_plot.legend()

#Gráfico año 2017
bar_plot = sns.barplot(x='Hombres', y='Edad', data=pop_2017, order=AgeClass, color='royalblue', lw=0, ax=axes[1], label="Hombres")
bar_plot = sns.barplot(x='Mujeres', y='Edad', data=pop_2017, order=AgeClass, color='tomato', lw=0, ax=axes[1], label="Mujeres")
axes[1].set_xlim(-800000,800000)
bar_plot.set_xticklabels(labels)
bar_plot.set_xlabel('')
bar_plot.set_title("Pirámide problacional año 2017")
bar_plot.legend()
plt.show()
```

<font color="orangered">Podemos conlcuir que la población en Chile para el año 2002 tiene una forma estacionaria a comportarse de forma regresiva en el año 2017. De rango de edad con mayor población paso de 10-14 años de edad en 2002 a 25-29 años en 2017, evidenciando que la población lentamente se esta avejentando.</font>

## <font color="orangered"> 2 - Analisis ganacias versus presupuesto peliculas Marvel </font>

```python
df_marvel = pd.read_excel("/content/drive/MyDrive/Analisis Exploratorio/Marvel.xlsx")
df_marvel['año'] = pd.DatetimeIndex(df_marvel['FechaLanzamientoEEUU']).year
df_marvel.head(5)
```

```python
##Se grafica la Matriz de correlacion.
sns.set(rc = {'figure.figsize':(9,6)})
# mask
mask = np.triu(np.ones_like(df_marvel.corr(), dtype=np.bool))
# adjust mask and df
mask = mask[1:, :-1]
corr = df_marvel.corr().iloc[1:,:-1].copy()

sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", 
           vmin=-1, vmax=1, cbar_kws={"shrink": .8})
sns.color_palette("ch:s=-.2,r=.6", as_cmap=False)
title = 'Matriz de correlacion - Peliculas de Marvel '
plt.title(title, loc='left', fontsize=14)
# yticks
plt.yticks(rotation=0)
plt.show()
```

<font color="orangered">**A partir de la matriz de correlacion se puede concluir lo siguiente:**

<font color="orangered">1) Existe un correlacion fuerte entre lo recaudado la primera semana de estreno en EEUU Canada con la recaudacion global, luego hay una alta correlacion entre eeuu y canada y el presupuesto, por lo tanto mientras mas alto sea el presupuesto ambas variables van aumentar.

<font color="orangered">2) La cantidad de premios Oscar ganados no se ven afectados ante un aumento en el presupuesto de la pelicula.

<font color="orangered">3) Existe una correlacion moderada entre el promedio de criticas y las nominaciones a los premios oscar

<font color="orangered">4) Es practicamente nula la correlacion entre los premios oscar ganados y la calificacion MDB (las mas importante BD cine y televisión del mundo).

<font color="orangered">5) La recaudacion de la primera semana EEUU Canada se mueve en direccion diferente a las nominaciones a los premios Oscar dado que da un correlacion negativa de -0.12
</font>

```python
fig3 = plt.figure(figsize=(21,10), constrained_layout=True)
gs = fig3.add_gridspec(3, 3)

f3_ax1 = fig3.add_subplot(gs[0, :])
f3_ax1.set_title('Recaudacion global por año')
f3_ax1.ticklabel_format(style='plain')
sns.barplot(x='año', y='RecaudacionGlobal', hue = 'Productor',data=df_marvel, ax = f3_ax1, palette='rocket')

f3_ax2 = fig3.add_subplot(gs[1, :-1])
f3_ax2.ticklabel_format(style='plain')
sns.lineplot(x='Genero', y='RecaudacionGlobal', hue = 'Productor', data=df_marvel, ci='sd', ax = f3_ax2, palette='rocket')

f3_ax3 = fig3.add_subplot(gs[1:, -1])
f3_ax3.ticklabel_format(style='plain')
sns.boxplot(x='Presupuesto', data=df_marvel, ax = f3_ax3,boxprops=dict(alpha=0.4), palette='rocket', orient='h')
sns.swarmplot(x='Presupuesto', data=df_marvel, ax = f3_ax3, palette='rocket')

f3_ax4 = fig3.add_subplot(gs[-1, 0])
f3_ax4.ticklabel_format(style='plain')
sns.regplot(x='Presupuesto', y='RecaudacionGlobal', data=df_marvel,  x_estimator=np.mean, scatter=True, color='crimson', line_kws={"color": "dimgray"},  ax = f3_ax4)

f3_ax5 = fig3.add_subplot(gs[-1, -2])
f3_ax5.ticklabel_format(useOffset=False, style='plain')
sns.regplot(x='RecaudacionEEUUCanadaPrimeraSemana', y='RecaudacionGlobal', data=df_marvel, ax = f3_ax5, color='crimson', line_kws={"color": "dimgray"})

plt.show()
```

<font color="orangered">**Segun la grafica de acuerdo a las variables, se infiere lo siguiente:**<br>
<font color="orangered">
1) La popularidad de Marvel aumenta en los años dado, que aumenta la recaudacion global desde el 2008 al 2019<br>
<font color="orangered">
2) El genero que mas atrae a los espectadores es accion, aventura y drama por que es el mas recaudacion globlal obtiene.<br>
<font color="orangered">
3) Los graficos de regresion confirman lo anteriormente visto en la matriz de correlacion, si el presupuesto aumenta la recaudacion global tambien y asi tambien se infiere que lo recaudacion en la primera semana EEUU y Canada tiene una correlacion fuerte con la recaudacion global.<br>
<font color="orangered">
4) Segun el grafico de bigotes, solamente dos peliculas salen del presupuesto que habitualmente se utiliza para este tipo de repartos.<br>
