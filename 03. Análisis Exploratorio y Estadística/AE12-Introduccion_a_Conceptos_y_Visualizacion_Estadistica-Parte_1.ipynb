**¿Qué es el Análisis Exploratorio de Datos?**

El __Análisis Exploratorio de Datos__ o (__EDA__) consiste en una serie de técnicas y herramientas que nos permiten comprender los conjuntos de datos resumiendo sus características principales, a menudo graficándolos visualmente y utilizando descriptores estadisticos. Este paso es muy importante especialmente cuando llegamos a modelar los datos para aplicar el aprendizaje automático. La visualización en EDA consta de histogramas, boxplots, diagrama de dispersión y muchos más. A menudo se necesita mucho tiempo para explorar los datos. A través del proceso de EDA, podemos definir el enunciado del problema, desarrollar nuestras preguntas de investigación, realizar el diseño experimental, entre otrsa cosas.

La **Estadística** es una disciplina científica que se ocupa de la obtención, ordenación  y análisis de un conjunto de datos con el fin de obtener explicaciones y predicciones sobre un fenómeno en estudio.



## <font color='blue'>**Conceptos básicos**</font>

**Población** : conjunto sobre el que estamos interesados en obtener conclusiones y  hacer inferencias.

**Muestra** : subconjunto de la población, sobre la cual se harán las observaciones  o mediciones.

**Variable**	: es una característica de un objeto o sujeto, que va variando entre los   diferentes individuos  de una población.

**Dato** : valores que se obtienen al realizar un estudio estadístico.

**Parámetro** : cantidad numérica que describe una característica de una Población.

**Estadístico** : cantidad numérica que describe una característica de una Muestra.

La **inferencia estadística** esta basada en el estudio de las **muestras**. La muestra debe ser **representativa de la
población** para extraer conclusiones validas sobre esta población. La muestra debe ser **aleatoria**.

## <font color='blue'>**Medidas de Tendencia Central**</font>

Las medidas de tendencia central son  valores que se encuentra hacia el centro de la distribución de los valores de una serie de observaciones.

La **Moda** de una variable Cualitativa o Cuantitativa, es el valor que se presenta con mayor frecuencia en una distribución de datos.

1. Cuando __dos datos__ se presentan con la misma frecuencia y ésta es la más alta, ambos datos son moda, es decir,  la distribución es bimodal.
2. Cuando __dos o más datos__ se presentan con la misma frecuencia y esta es la más alta, todos datos son moda, es decir, la distribución es multimodal.
3. Cuando __todos los datos__ tienen la misma frecuencia, se dice que no hay moda.

La **Mediana** de una variable cualitativa ordinal o cuantitativa, corresponde al valor intermedio, cuando los valores de la variable se presentan en orden de magnitud creciente (o decreciente).


$$ Me = X_{\frac{n}{2}}, \mbox{  Para el caso que n es impar.}  $$ 

$$ Me = \frac{X_{\frac{n}{2}} + X_{\frac{n}{2}+1}}{2},  \mbox{  Para el caso que n es par.}$$ 

La **media aritmetica** de una variable cuantitativa, se calcula sumando los valores y dividiéndolos por el total de ellos.

$$ Media = \frac{\sum X_i}{n}$$

**Rango**: No es una medida significativa, pues solo indica cuán dispersos están los datos entre los extremos. 

$$R = X_{max} - X_{min}$$

**Coeficiente de variación**: También llamado _Coeficiente de Variación de Pearson_, es una medida estadística que nos informa acerca de la dispersión relativa de un conjunto de datos. Es decir, nos informa, al igual que otras medidas de dispersión, si una variable se mueve mucho, poco, más o menos que otra. Se utiliza cuando se necesita comparar la dispersión de distribuciones de variables que tienen distintas unidades de medida. 

$$CV = \frac{S}{\bar{X}}$$

Su fórmula expresa la desviación estándar como porcentaje de la media aritmética, mostrando una interpretación relativa del grado de variabilidad, independiente de la escala de la variable, a diferencia de la desviación típica o estándar. Por otro lado presenta problemas ya que a diferencia de la desviación típica este coeficiente es fuertemente sensible ante cambios de origen en la variable. Por ello es importante que todos los valores sean positivos y su media dé, por tanto, un valor positivo. A mayor valor del coeficiente de variación mayor heterogeneidad de los valores de la variable; y a menor C.V., mayor homogeneidad en los valores de la variable. Por ejemplo, si el C.V es menor o igual al 30%, significa que la media aritmética es representativa del conjunto de datos, por ende el conjunto de datos es "Homogéneo". ## <font color='blue'>**Medidas de posición**</font>

**Cuartiles**  son medidas de posición  que dividen  la distribución ordenada en cuatro partes iguales, ellos son $Q_1 , Q_2 , Q_3 $.

$Q_1$: ***Cuartil 1***. Es el valor de la variable que deja a la izquierda el 25% de los datos.

**Rango intercuartilico (IQR)**: es una estimación estadística de dispersión  de una distribución de datos y corresponde a:  

$$IQR = Q_3-Q_1$$

**Percentiles** son medidas de posición que dividen la distribución ordenada en 100 partes iguales, ellos son $P_1$, $P_2$, ..., $P_k$ , con $k=[1,99]$.

$P_K$ :  ***Percentil k***. Es el valor de la variable que deja a la izquierda el $k\%$ de los datos.

### Parámetros (poblacionales)
Dijimos que un *parámetro* es un valor que caractariza un aspecto de la población; e.g., media poblacional, mediana poblacional, desviación estándar poblacional, etc.

Analicemos los parámetros de la población **poblacion**. Como los datos no están guardados en una estructura Pandas (Serie, DataFrame), utilizaremos las librerías _Nympy_ y _scipy_ para calcular parámetros y estadísticos.

`````python
from scipy import stats # importando scipy.stats

print(f'Tamaño población\t\t= {len(poblacion):d} individuos')
print(f'Media poblacional\t\t= {poblacion.mean():4.2f}')
print(f'Desviación estándar poblacional\t= {poblacion.std():4.2f}')
print(f'Mediana poblacional\t\t= {np.median(poblacion):4.2f}')
`````

Ahora analicemos los parámetros de la población **edades**. Incorporamos una nueva medida, la __moda__. La moda es el valor que aparece con mayor frecuencia dentro de un conglomerado. En un grupo puede haber una única moda (`unimodal`), dos modas (`bimodal`), o más de dos modas (`multimodal`). Se llama a un grupo `amodal` cuando en él no se repiten los valores.

Por último, se conoce como `moda adyacente` cuando dos valores continuos tienen la misma cantidad de repeticiones. En este caso se saca el promedio de ambos.

`````python
print(f'Tamaño población edades\t\t\t= {len(edades):d} individuos')
print(f'Media de edades\t\t\t\t= {edades.mean():4.2f}')
print(f'Desviación estándar de las edades\t= {edades.std():4.2f}')
print(f'Mediana poblacional de las edades\t= {np.median(edades):4.2f}')
print(f'Moda poblacional es {stats.mode(edades)[0][0]} y se repite {stats.mode(edades)[1][0]} veces')
`````

### Estadísticos (muestrales)
Medida descriptiva de la muestra que se utiliza para estimar al respectivo parámetro poblacional. El *estadístico* es básicamente lo mismo que el *parámetro poblacional*, pero se diferencia en que se calcula a partir de la muestra y no de la población. <br>
La desvisción estándar para muestras se denomina "__desviación estándar insesgada__" ya que se calcula dividiendo por $(n - 1)$.

**Observación**

La desviación estándar muestral se denota con el símbolo: $S$ o $\widehat{\sigma}$

La media muestral se denota con el símbolo: $\overline{X}$ o $\widehat{\mu}$

Es la diferencia entre el valor del **parámetro poblacional** y el producido por el **estadístico muestral**.

### Gráficos de barras
Un diagrama de barras es una forma de representar datos donde la *longitud* de las barras representa la magnitud / tamaño de la característica / variable. Los gráficos de barras generalmente representan variables numéricas y categóricas agrupadas en intervalos.

Para crear un diagrama de barras (__bar plot__), podemos pasar uno de dos argumentos a través del parámetro `kind` en el método `plot()`:

* `kind=bar` Crea verticales.
* `kind=barh` Crea horizontales.

`````python
# Paso 1: Obtengamos los datos. 
years = list(range(1980, 2014))

df_iceland = df_canada.loc['Iceland', years] #todos los años para iceland
df_iceland.plot(kind='bar', figsize=(12, 4))

plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.title('Inmigrantes desde Islandia a Canadá desde 1980 a 2013')

plt.show()
`````

`````python
df_iceland.plot(kind='bar', figsize=(12, 4), rot=90) # girar los ticks 90 grados

plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.title('Inmigrantes desde Islandia a Canadá desde  1980 a 2013')

# Annotate arrow
plt.annotate('',                      # s: str. Lo dejará en blanco si no hay texto
             xy=(32.5, 70),             # coloque la punta de la flecha en el punto (año 2012, número de inmigrantes 70)
             xytext=(31.5, 45),         # coloque la base de la flecha en el punto (año 2008, número de inmigrantes 20)
             xycoords='data',         # utilizará el sistema de coordenadas del objeto que se anotará 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

plt.annotate('Post Crisis', # Texto para desplegar
             xy=(30.8, 50),                    # comenzar el texto en un punto (año 2008, número de inmigrantes 30)
             rotation=75,                  # basado en prueba y error para que coincida con la flecha
             va='bottom',                    # quiere que el texto esté alineado verticalmente en la parte inferior
             ha='left',                      # desea que el texto esté alineado horizontalmente a la izquierda.
            )

plt.show()
`````

### Graficos de Torta

Un "Grafico de torta" (__Pie Chart__) es un gráfico que muestra proporciones numéricas dividiendo un círculo (o gráfico circular) en sectores proporcionales. Lo más probable es que ya esté familiarizado con los gráficos circulares, ya que es ampliamente utilizado en negocios y medios. Podemos crear gráficos circulares en Matplotlib pasando la palabra clave `kind = pie`.

Usemos un gráfico circular para explorar la proporción (porcentaje) de nuevos inmigrantes agrupados por continentes para todo el período comprendido entre 1980 y 2013.

`````python
# Agrupemos por continentes y aplicamos suma. En el groupby estamos realizando el split por continente.
df_continents = df_canada.groupby('Continente', axis=0).sum()

# Nota: El resultado del metodo groupby es un objeto de tipo `groupby'.
# no podemos usarlo hasta que apliquemos una función (por ejemplo, suma ())
print(type(df_canada.groupby('Continente', axis=0)))
`````

`````python
df_continents.head()
`````

`````python
df_continents['Total'].plot(kind='pie',
                            figsize=(7, 8),
                            autopct='%1.1f%%', # Incorpora el porcentage. 
                            startangle=-45,     # Empieza con angulo 90° (Africa)
                            shadow=False,      # agregar sombra
                            )

plt.title('Inmigración a Canadá por continente')
plt.axis('equal') # Configura el gráfico para que parezca un círculo.
plt.show()
`````

### Box Plots

Un "Box plot" es una forma de representar estadísticamente la *distribución* de los datos a través de cinco dimensiones principales:

- **Minimun:** Número más pequeño en el conjunto de datos.
- **First quartile:** Número medio entre el `mínimo` y la` mediana`.
- **Second quartile (Median):** Número medio del conjunto de datos (ordenado).
- **Third quartile:** Número medio entre `mediana` y` máximo`.
- **Maximum:** Número más alto en el conjunto de datos.
