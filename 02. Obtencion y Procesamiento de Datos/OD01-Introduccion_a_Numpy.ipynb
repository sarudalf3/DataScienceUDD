## <font color='blue'>**Librería NumPy**</font>

**NumPy** es la librería por excelencia para computación científica en Python. Integra muchas funciones de cálculo matricial de N dimensiones, y una variedad de rutinas para operaciones rápidas con matrices, que incluyen manipulación matemática, lógica, de tamaño y forma, clasificación, selección, I/O, transformada de Fourier, algebra lineal, operaciones estadísticas, simulación aleatoria y muchas otras.

<img src='https://drive.google.com/uc?export=view&id=1_foOq3ZJE-GKzm9hDpuTDrT8c_YxkorY' width="400" align="center" style="margin-right: 20px">

En el núcleo del paquete NumPy está el objeto **ndarray**. Este encapsula matrices n-dimensionales de tipos de datos homogeneos, con muchas operaciones que se realizan en código compilado (lenguaje C) para mejorar el rendimiento. Es la librería de cálculo más popular debido a su facilidad de uso y la rapidez de sus cálculos.

Existen varias diferencias importantes entre las matrices de NumPy y las secuencias estándar de Python:

* Las matrices NumPy tienen un tamaño fijo al momento de la creación, a diferencia de las listas de Python (que pueden crecer dinámicamente). Cambiar el tamaño de un ndarray creará una nueva matriz y eliminará la original.


* Se requiere que todos los elementos de una matriz NumPy correspondan al mismo tipo de datos y, por lo tanto, tendrán el mismo tamaño en la memoria. La excepción: uno puede tener matrices de objetos (Python, incluido NumPy), lo que permite matrices de elementos de diferentes tamaños.


* Las matrices NumPy facilitan las operaciones matemáticas avanzadas (entre otras), en grandes cantidades de datos. Por lo general, estas operaciones se ejecutan de manera más eficiente y con menos código que al usar las secuencias integradas de Python.


* Una multitud de paquetes científicos y matemáticos basados en Python utilizan matrices NumPy; aunque estos paquetes suelen admitir la entrada de secuencias estándar de Python, convierten dicha entrada en matrices NumPy antes del procesamiento y, a menudo, generan matrices NumPy. En otras palabras, para usar de manera eficiente gran parte (quizás incluso la mayoría) del software científico/matemático basado en Python de hoy en día, simplemente saber cómo usar los tipos de secuencia incorporados de Python es insuficiente; uno también necesita saber cómo usar las matrices NumPy.

### <font color='blue'>**¿Cuál es la diferencia entre una lista y un matriz NumPy?**</font>

NumPy ofrece una enorme variedad de formas rápidas y eficientes de crear matrices y manipular datos numéricos. Si bien una lista de Python puede contener diferentes tipos de datos dentro de una sola lista, todos los elementos de una matriz NumPy deben ser homogéneos. Las operaciones matemáticas que deben realizarse en matrices serían extremadamente ineficaces si las matrices no fueran homogéneas.

#### <font color='blue'>**¿Por qué usar NumPy?**</font>

Las matrices NumPy son más rápidas y compactas que las listas de Python. Una matriz consume menos memoria. NumPy usa mucha menos memoria para almacenar datos y proporciona un mecanismo para especificar los tipos de datos. Esto permite optimizar aún más el código.

````python
import sys

S = range(10000)
print('Resultado de memoria asignada a la lista de Python:')
print(sys.getsizeof(5) * len(S))
print()
D = np.arange(10000)
print('Resultado de memoria asignada a NumPy array:')
print(D.size * D.itemsize)
````

````python
import time
SIZE = 1000000

L1 = range(SIZE)
L2 = range(SIZE)
A1 = np.arange(SIZE)
A2 = np.arange(SIZE)

start = time.time()
result = [(x,y) for x,y in zip(L1, L2)]
print('Resultado test de velocidad para lista de Python [ms]:')
print((time.time() - start) * 1000)
print()

start = time.time()
result = A1 + A2
print('Resultado test de velocidad para NumPy array [ms]:')
print((time.time() - start) * 1000)
````

<font color="orange">Lo visto muestra una eficiencia en tiempo y en recursos al usar numpy</font>

#### <font color='darkorange'>A continuación se muestra la comparación de la memoria y tiempo utilizado para concatenar 2 listas o arrays. En ambos casos se observa una diferencia muy grande que se acentúa mientras más repeticiones hay.</font>

````python
import matplotlib.pyplot as plt
def comparacion_memoria(size, n=2):
    """
    Función que compara la memoria utilizada por listas y arrays para distintos tamaños.
    inputs:
        size: tamaño de la lista y array.
        n: número de calculos de memoria (se repite el calculo de memoria n veces para promediar)
    output:
        devuelve el promedio dememoria utilizada en la creación de una lista y un array.
    """
    result = []
    L1 = range(size)
    L2 = range(size)
    A1 = np.arange(size)
    A2 = np.arange(size)

    mem_listas = 0
    for tl in range(n):
        lista = [(x,y) for x,y in zip(L1, L2)]
        #mem_listas += sys.getsizeof(5) * len(lista)
        mem_listas += sum([sys.getsizeof(i) for i in lista])/1048576 # transforma a MB

    result = result + [mem_listas/n]

    mem_arrays = 0
    for tl in range(n):
        start = time.time()
        A = A1 + A2
        #mem_arrays += A.size * A.itemsize / 1048576 # transforma a MB
        #mem_arrays += sys.getsizeof(A)/1048576 # transforma a MB
        mem_arrays += A.nbytes/1048576 # otra función de numpy, devuelve los bytes de todo el objeto

    result = result + [mem_arrays/n]
    return result

compara_memoria = [comparacion_memoria(i, 5) for i in range(1000,1001001,200000)]
plt.plot([j[0] for j in compara_memoria], label="Listas")
plt.plot([j[1] for j in compara_memoria], label="Arrays")
plt.title("Comparación de memoria entre listas y np.array [MB]")
plt.legend()
plt.show()
````

<font color="red">Claramente la diferencia de memoria se acrecienta a la vez que el número de cálculos aumenta</font>

````python
def comparacion_tiempo(size, n=2):
    """
    Función que compara el tiempo de ejecución de listas vs arrays para distintos tamaños.
    inputs:
        size: tamaño de la lista y array.
        n: número de calculos de tiempo (se repite el calculo de tiempo n veces para promediar)
    output:
        devuelve el promedio del tiempo de ejecución de la creación de una lista y un array.
    """
    result = []
    L1 = range(size)
    L2 = range(size)
    A1 = np.arange(size)
    A2 = np.arange(size)

    time_listas = 0
    for tl in range(n):
        start = time.time()
        lista = [(x,y) for x,y in zip(L1, L2)]
        time_listas += (time.time() - start) * 1000

    result = result + [time_listas/n]

    time_arrays = 0
    for tl in range(n):
        start = time.time()
        lista = A1 + A2
        time_arrays += (time.time() - start) * 1000

    result = result + [time_arrays/n]
    return result

compara_tiempo = [comparacion_tiempo(i, 5) for i in range(1000,1001001,200000)]

plt.plot([j[0] for j in compara_tiempo], label="Listas")
plt.plot([j[1] for j in compara_tiempo], label="Arrays")
plt.title("Comparación de velocidad entre listas y np.array [ms]")
plt.legend()
plt.show()
````

<font color="red">El tiempo comparado entre listas y objetos ndarray es más amplio que la memoria, ya que los elementos ndarray toman muy poco tiempo en ser ejecutados</font>
